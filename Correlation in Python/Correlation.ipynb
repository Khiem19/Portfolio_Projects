```python
#Import Libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)


df = pd.read_csv(r'movies.csv')
```

```python
df.head()
```

```python
#check for missing data
for col in df.columns:
    print(f'{col} - Missing values: {df[col].isnull().sum()}')
    
    
#Fill nulls with 0
df.fillna(0, inplace=True)
```

```python
#Data types for columns

df.dtypes
```

```python
#Change datatype of columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')

df.dtypes
```

```python
#create corrected year column
df['yearcorrect'] = df['released'].str.extract(r'(\d{4})')

df
```

```python
df.sort_values(by = ['gross'], inplace = True, ascending = False)

df
```

```python
# pd.set_option('display.max_rows', None)
# pd.reset_option('display.max_rows')
```

```python
#Drop any duplicates

# df['company'].drop_duplicates().sort_values(ascending=False)
df.drop_duplicates()
```

```python
#Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for films')

plt.show()
```

```python
df.head()
```

```python
#Budget vs Gross in seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color" : "red"}, line_kws= {"color" : "blue"})
```

```python
#Correlation
df.corr(numeric_only=True)
```

```python
correlation_matrix = df.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
#Looks at Company

df_numerized = df.copy()

for col_name in df_numerized.columns:
    if df_numerized[col_name].dtype == 'object':
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

# df_numerized.dtypes
df_numerized
```

```python
df
```

```python
correlation_matrix = df_numerized.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```
